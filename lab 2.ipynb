```python
import numpy as np
a=np.array([2,4,1,6,9,6,7])
d=np.array([2,4,1,7,9,4,5])
b=np.mean(a)
print("mean=",b)
c=np.median(a)
print("median=",c)
print("varience=",np.var(a))
print("standard deviation=",np.std(a))
print("sorted array=",np.sort(a))
print("sum of array=",np.sum(a))
print("addition of 2 array=",np.add(a,d))
print("condition:",np.where(a>4))
```

```python
from matplotlib import pyplot as plt
% matplotlib inline
import numpy as np
x=[1,2,4,3,6,8]
y=np.sin(x)
plt.plot(x,y,'b')
plt.show()
plt.bar(x,y)
plt.show()
```

```python
from matplotlib import pyplot as plt
x=[2,6,3,9]
y=[4,7,5,9]
plt.plot(x,'y')
plt.plot(y,'b')
plt.title("AXIS")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend(['x','y'])
plt.show()
```

```python

from matplotlib import pyplot as plt
x=[2,6,3,9,3,6,8,1]
y=[4,7,5,9,5,8,3,9]
plt.plot(x,'y*')
plt.plot(y,'b+')
plt.title("AXIS")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend(['x','y'])
plt.show()
```

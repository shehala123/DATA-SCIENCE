```python
#Decision tree algorthm

#import
from sklearn import tree #classifier
from sklearn.model_selection import train_test_split #to split data 
from sklearn.datasets import load_iris #forn data sets(Iris)
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
#load dataset
irisData=load_iris ()
print("\n------------------\ndata and targets\n------------------\n")
print(irisData)
#create feature and target arrays
print("\n------------------\ndata \n------------------\n")
x=irisData.data
print(x)
print("\n------------------\ntargets\n------------------\n")
y=irisData.target
print(y)
#split data to training and testing data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print("\n------------------\ntrain data\n------------------\n")
print(x_train)
print("\n------------------\ntest data\n------------------\n")
print(x_test)
print("\n------------------\ntrain target\n------------------\n")
print(y_train)
print("\n------------------\ntest target\n------------------\n")
print(y_test)
#traindata+algorthm=model
clt=tree.DecisionTreeClassifier(criterion='entropy')
clt=clt.fit(x_train,y_train)
#predict the future
print("\n------------------\npredict test data target\n------------------\n")
print(clt.predict(x_test))
#test the accurence

print("\n------------------\naccurency of test data and target\n------------------\n")
score=accuracy_score(clt.predict(x_test),y_test)
print(score)
print("\n------------------\ndecision tree\n------------------\n")
plt.figure(figsize=(12,12))
tree.plot_tree(clt)
plt.show()
```

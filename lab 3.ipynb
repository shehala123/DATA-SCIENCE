```python
def add(a,b):
    c=a+b
    print("result is:",c)
def sub(a,b):
    c=a-b
    print("result is:",c)
def mul(a,b):
    c=a*b
    print("result is:",c)
def div(a,b):
    c=a/b
    print("result is:",c)

a=int(input("enter first number:"))
b=int(input("enter second number:"))
print("enter your choice:")
d=int(input(" 1.Addition\n 2.substraction\n 3. multiplication\n 4.division\n"))
if(d==1):
    add(a,b)
elif(d==2):
    sub(a,b)
elif(d==3):
    mul(a,b)
elif(d==4):
    div(a,b)
else:
    print("invalid choice")
```

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
x=np.array([2,6,3,9])
y=np.sin(x)
z=np.cos(x)
plt.grid()
plt.plot(x,y)
plt.plot(x,z)
plt.title("GRAPH")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend(['sin','cos'],loc='upper right')
plt.show()
```

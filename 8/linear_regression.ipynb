```python
#KNN algorthm

#import
from sklearn.linear_model import LinearRegression#classifier
from sklearn.model_selection import train_test_split #to split data 
from sklearn.datasets import load_iris #forn data sets(Iris)
from sklearn.metrics import mean_squared_error,r2_score
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
#load dataset
irisData=load_iris ()
print("\n------------------\ndata and targets\n------------------\n")
print(irisData)
#create feature and target arrays
print("\n------------------\ndata \n------------------\n")
x=irisData.data
print(x)
print("\n------------------\ntargets\n------------------\n")
y=irisData.target
print(y)
#split data to training and testing data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print("\n------------------\ntrain data\n------------------\n")
print(x_train)
print("\n------------------\ntest data\n------------------\n")
print(x_test)
print("\n------------------\ntrain target\n------------------\n")
print(y_train)
print("\n------------------\ntest target\n------------------\n")
print(y_test)
#traindata+algorthm=model
model=LinearRegression()
model.fit(x_train,y_train)
#y itercept
print("\n------------------\ny itercept\n------------------\n")
print(model.intercept_)
#predict the future
print("\n----------------------------------\npredict target and print actual target of y\n\n----------------------------------\n\n")
print(model.predict(x_test))
print("\n\n")
print(y_test)
#test the accurence
print("\n------------------\naccurency of train data and target\n------------------\n")
print(model.score(x_train,y_train))
print("\n------------------\naccurency of test data and target\n------------------\n")
print(model.score(x_test,y_test))
print("\n-----------------------------\nperformence of model using mean_squared_error\n----------------------------\n")
y_predict=model.predict(x_test)
print(mean_squared_error(y_test,y_predict))
print("\n-----------------------------\nperformence of model using r2_score\n----------------------------\n")
print(r2_score(y_test,y_predict))


print("\n--------------------------------------------\ngraph showing the variation between y_test and y_predict\n---------------------------------------------\n")
plt.plot(y_test)
plt.plot(y_predict)
plt.show()
```

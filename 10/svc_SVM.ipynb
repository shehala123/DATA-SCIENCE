```python
#svc
from sklearn.model_selection import train_test_split #to split data 
from sklearn.datasets import load_iris #forn data sets(Iris)
from sklearn.svm import SVC
from sklearn import svm
#load dataset
irisData=load_iris ()
print("\n------------------\ndata and targets\n------------------\n")
print(irisData)
#create feature and target arrays
print("\n------------------\ndata \n------------------\n")
x=irisData.data
print(x)
print("\n------------------\ntargets\n------------------\n")
y=irisData.target
print(y)
#split data to training and testing data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print("\n------------------\ntrain data\n------------------\n")
print(x_train)
print("\n------------------\ntest data\n------------------\n")
print(x_test)
print("\n------------------\ntrain target\n------------------\n")
print(y_train)
print("\n------------------\ntest target\n------------------\n")
print(y_test)
#traindata+algorthm=model
svc_classifier = svm.SVC(kernel='linear',random_state=0 ).fit(x_train,y_train)
#predict the future
print("\n-----------------------------------\npredict test data target and actual target\n---------------------------------\n")
print(svc_classifier.predict(x_test))
print("\n")
print(y_test)
#test the accurence
print("\n------------------\naccurency of train data and target\n------------------\n")
print(svc_classifier.score(x_train,y_train))
print("\n------------------\naccurency of test data and target\n------------------\n")
print(svc_classifier.score(x_test,y_test))
```

```python
#SVM
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
from sklearn import svm
import matplotlib.pyplot as plt
iris = load_iris()
X = iris.data[:, :2]
y = iris.target
print(X)
print(y)
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
print(x_min, x_max)
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
print(y_min, y_max)
h = (x_max / x_min)/100
print(h)
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
print(xx,yy)
X_plot = np.c_[xx.ravel(), yy.ravel()]
svc_classifier = svm.SVC(kernel='linear', C=1.0).fit(X, y)
Z = svc_classifier.predict(X_plot)
Z = Z.reshape(xx.shape)
plt.figure(figsize=(15, 5))
plt.subplot(121)
plt.contourf(xx, yy, Z, cmap=plt.cm.tab10, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.title('Support Vector Classifier with linear kernel')
```

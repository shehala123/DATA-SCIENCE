```python
#rotate

from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
from PIL import Image
from scipy import ndimage
a=Image.open("123.jpeg")
b=np.array(a)
c=ndimage.rotate(b,50,cval=128)
d=Image.fromarray(c)
plt.imshow(d)
```

```python
#scaling

from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
from PIL import Image
from scipy import ndimage
a=Image.open("123.jpeg")
b=np.array(a)
c=ndimage.zoom(b,(0.25,1,1))
d=Image.fromarray(c)
plt.imshow(d)
```

```python
#shear

from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
from PIL import Image
from scipy import ndimage
a=Image.open("123.jpeg")
b=np.array(a)
h,w,s=b.shape
d=[[1,0,0],[0.5,1,0],[0,0,1]]

shear=ndimage.affine_transform(b,d)
d=Image.fromarray(shear)
plt.imshow(d)

```
